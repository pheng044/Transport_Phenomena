## Appendix 1: Mathematical Operators in Cartesian Coordinates

### Vector Calculus Review

The following a quick review of common operations from multivariable
calculus which are useful in deriving conservation equations. We will
only consider Cartesian coordinates as operators in cylindrical and
spherical coordinates become *much* more complex. We let $f$ be any
function, $\mathbf{a}$ and $\mathbf{b}$ be vectors, $\mathbf{F}$ be a
vector field, and $\mathbf{P}$ be any differentiable quantity, 

$$\ 
f=f(x,y,z),\quad
\mathbf{a}=
\begin{pmatrix}
a_1 \\
a_2 \\
a_3 \\
\end{pmatrix},\quad
\mathbf{b}=
\begin{pmatrix}
b_1 \\
b_2 \\
b_3 \\
\end{pmatrix},\quad
\mathbf{F}=
\begin{pmatrix}
f(x,y,z) \\
g(x,y,z) \\
h(x,y,z) \\
\end{pmatrix}.$$ 

The dot product between two vectors $\mathbf{a}$ and
$\mathbf{b}$, denoted as $\mathbf{a}\cdot\mathbf{b}$, is defined as,

$$\ \mathbf{a}\cdot\mathbf{b}:=\left|\mathbf{a}\right|\left|\mathbf{b}\right|\cos{\left(\theta\right)}=a_1b_1+a_2b_2+a_3b_3.$$

Geometrically, the dot product is the projection, or amount of one
vector in the direction of another vector. Two vectors are said to be
perpendicular (orthogonal) when $\mathbf{a}\cdot\mathbf{b}=0$ since
$\cos{\left(\theta\right)}=0$ when $\theta=\pi/2$.

The gradient of a function, $f$, denoted as $\nabla f$, is defined as,

$$\ \nabla f :=
\begin{pmatrix}
\displaystyle \frac{\displaystyle \partial^{}f}{\displaystyle \partial x ^{}} \\
\displaystyle \frac{\displaystyle \partial^{}f}{\displaystyle \partial y ^{}} \\
\displaystyle \frac{\displaystyle \partial^{}f}{\displaystyle \partial z ^{}} \\
\end{pmatrix}.\quad$$ 

Geometrically, the gradient is normal to curves of
constant $f$, that is, if $\mathbf{dx}$ is a small change in position
such that the output of $f$ remains constant, then,

$$\ \nabla f\cdot\mathbf{dx}=0.$$ 

Conversely, let $\mathbf{v}$ be any
unit vector ($\left|\mathbf{v}\right|=1$), then, the directional
derivative is defined as,

$$\ \nabla f\cdot\mathbf{v}:=\left|\nabla f\right|\left|\mathbf{v}\right|\cos{\left(\theta\right)}=\left|\nabla f\right|\cos{\left(\theta\right)}.$$

The directional derivative measures the change of a function when moving
in a direction, $\mathbf{v}$. This quantity is maximized when
$\theta = 0$ where cosine is equal to 1. This means that if $\mathbf{v}$
points in the gradient direction, then the directional derivative, or
the change in $f$, is the greatest. This is often stated as the gradient
vector points in the steepest ascent direction. By using the same line
of reasoning when $\theta = \pi$, we see that the negative gradient
points in the steepest descent direction.

The divergence of a vector field, $\mathbf{F}$, denoted as
$\nabla\cdot\mathbf{F}$, is defined as,

$$\ \nabla\cdot\mathbf{F} := \displaystyle \frac{\displaystyle \partial^{}f}{\displaystyle \partial x ^{}} + \displaystyle \frac{\displaystyle \partial^{}g}{\displaystyle \partial y ^{}} + \displaystyle \frac{\displaystyle \partial^{}h}{\displaystyle \partial z ^{}} = \left(\displaystyle \frac{\displaystyle \partial^{}}{\displaystyle \partial x ^{}},\ \displaystyle \frac{\displaystyle \partial^{}}{\displaystyle \partial y ^{}},\ \displaystyle \frac{\displaystyle \partial^{}}{\displaystyle \partial z ^{}}\right)\cdot\left(f,\, g,\, h\right).$$

Physically, the divergence represents the flux of $\mathbf{F}$ at a
point or how much of $\mathbf{F}$ is coming in or out of a point. To see
this, draw a box with side lengths $\Delta x$, $\Delta y$, and
$\Delta z$, then the flux of $\mathbf{F}$, denoted as $\varphi$, can be
evaluated as,

$$\ \varphi \Delta x\Delta y\Delta z = \left(f|_{x+\Delta x}-f|_{x}\right)\Delta y\Delta z + \left(g|_{y+\Delta y}-g|_{y}\right)\Delta x\Delta z + \left(h|_{z+\Delta z}-h|_{z}\right)\Delta x\Delta y,$$

dividing by $\Delta x\Delta y\Delta z$,

$$\ \varphi = \frac{f|_{x+\Delta x}-f|_{x}}{\Delta x} + \frac{g|_{y+\Delta y}-g|_{y}}{\Delta y} + \frac{h|_{z+\Delta z}-h|_{z}}{\Delta z},$$

letting $\Delta x$, $\Delta y$, and $\Delta z$ approch 0, we get,

$$\ \varphi =  \displaystyle \frac{\displaystyle \partial^{}f}{\displaystyle \partial x ^{}} + \displaystyle \frac{\displaystyle \partial^{}g}{\displaystyle \partial y ^{}} + \displaystyle \frac{\displaystyle \partial^{}h}{\displaystyle \partial z ^{}}.$$

### The Laplacian

The Laplacian is defined as,

$$\nabla^2\mathbf{P}:=\nabla\cdot \nabla\mathbf{P}=\displaystyle \frac{\displaystyle \partial^{2}\mathbf{P}}{\displaystyle \partial x ^{2}}+\displaystyle \frac{\displaystyle \partial^{2}\mathbf{P}}{\displaystyle \partial y ^{2}}+\displaystyle \frac{\displaystyle \partial^{2}\mathbf{P}}{\displaystyle \partial z ^{2}}.$$

### The Convective Derivative

The convective derivative is defined as,

$$\ \mathbf{u}\cdot\nabla\mathbf{P} := u_x\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial x ^{}}+u_y\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial y ^{}}+u_z\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial z ^{}},$$

where $\mathbf{u}$ is the velocity of the fluid carrying the quantity
$\mathbf{P}$.

### The Material Derivative

The material derivative is defined as, 

$$\begin{aligned}
    \dfrac{D \mathbf{P}}{Dt}&:=\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial t ^{}}+\mathbf{u}\cdot\nabla \mathbf{P}\\
    \ &=\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial t ^{}}+u_x\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial x ^{}}+u_y\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial y ^{}}+u_z\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial z ^{}}\\
    \ &=\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial t ^{}}+\displaystyle \dfrac{d^{}x}{dt^{}}\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial x ^{}}+\displaystyle \dfrac{d^{}y}{dt^{}}\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial y ^{}}+\displaystyle \dfrac{d^{}z}{dt^{}}\displaystyle \frac{\displaystyle \partial^{}\mathbf{P}}{\displaystyle \partial z ^{}}.
\end{aligned}$$
