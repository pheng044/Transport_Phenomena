# Developing the Transport Equations (II)

## Putting it All Together

We have developed flux and generation relations for mass, momentum, and
energy transport. We also have a general conservation equation which
relates these fluxes and generation terms, so lets put them together! To
get started, an important definition is the so-called \"Laplace
operator\" or \"Laplacian\", which is a generalization of the second
derivative,

$$\nabla\cdot \nabla=\nabla^2=\underbrace{\displaystyle \frac{\displaystyle \partial^{2}}{\displaystyle \partial x ^{2}}+\displaystyle \frac{\displaystyle \partial^{2}}{\displaystyle \partial y ^{2}}+\displaystyle \frac{\displaystyle \partial^{2}}{\displaystyle \partial z ^{2}}}_{\mathrm{Cartesian}}.$$

Physically, this operator represents the diffusion or \"spreading out\"
of the quantity it acts on. For example, the figure below represents the
diffusion, $\nabla^2\, T$, of temperature, $T$, over time in
non-dimensionalized units.

```python
from matplotlib import pyplot as plt
plt.imshow(plt.imread('Images/Heat_Eqn_Gaussian_IC.jpg'))
plt.axis('off')
plt.show()
```

We will see that diffusion has to due with microscopic transport and
arises out of the non-convective flux terms from the conservation
equations. Thus, diffusion captures our fundamental principle that
nature seeks to spread out high energy states.

### Mass Conservation Equation

Beyond this point, putting everything together is just mathematical
manipulation. It is beneficial to see this once, but its not completely
necessary to understand every mathematical step to understand the
physical meaning of the equations. It is more important to note the
[assumptions]{.underline} in deriving the equations and *when* those
assumptions are imposed. We'll start with the general mass conservation
equation. Recall the general conservation equation,

$$\ \rho\mathbf{g}=\nabla\cdot \left(\boldsymbol{\varphi}+\rho\mathbf{b}\mathbf{v}\right)+\displaystyle \frac{\displaystyle \partial^{}(\rho\mathbf{b})}{\displaystyle \partial t ^{}}.$$

Our conserved intensive quantity is the concentration,
$\rho\mathbf{b}=C$. To avoid specific reaction (generation) terms, let
$R$ be the generation of a substance, with concentration, $C$, i.e.
$\rho\mathbf{g}=R$. From Fick's Law, we also know
$\boldsymbol{\varphi}=-D\nabla C$. Putting it all together,

$$\ R= \nabla\cdot \left(-D\nabla C+\mathbf{v}C\right)+\displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}}.$$

Expanding out the flux terms,

$$\ \underbrace{R}_{\mathrm{Generation}} = \underbrace{-\nabla\cdot \left(D\nabla C\right)}_{\mathrm{Diffusion}}\ +\ \underbrace{\nabla\cdot \left(\mathbf{v}C\right)}_{\mathrm{Convective\ Flux}}\quad+\underbrace{\displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}}}_{\mathrm{Accumulation}}.$$

This is can also be written in terms of the material derivative defined
as,

$$\ \dfrac{D C}{Dt}=\displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}}+\mathbf{v}\cdot\nabla C.$$

A more visual explanation of the material derivative is given in
Appendix 1. With the material derivative, the general mass conservation
equation becomes,

$$\ R= -\nabla\cdot \left(D\nabla C\right)+C\nabla\cdot \mathbf{v}+\dfrac{D C}{Dt}.$$

Often, $D$ is constant, giving,

$$\ R= -D\nabla^2 C+C\nabla\cdot \mathbf{v}+\dfrac{D C}{Dt}.$$ 
If the fluid is assumed incompressible $(\nabla\cdot \mathbf{v}=0)$, this
gives, 

$$\  R + D\nabla^2 C = \dfrac{D C}{Dt},$$ 

or,

$$\ \underbrace{\displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}}}_{\mathrm{Accumulation}}\ +\quad\underbrace{ \mathbf{v}\cdot\nabla C}_{\mathrm{Convection}} \quad=\quad \underbrace{D\nabla^2 C}_{\mathrm{Diffusion}} \ \ \ +\  \underbrace{R}_{\mathrm{Generation}}.$$

The differential equation has the same interpretation as the integral
conservation equation but with the non-convective flux term replaced
with diffusion; this is a common theme among transport equations. In
other words, differential conservation equations often follow,

$$\ \mathrm{ \{ Acummulation \} + \{ Convection \} = \{ Diffusion \} + \{ Generation \} }.$$

The accumulation term can also be interpreted as the \"local\" change in
time of the quantity of interest.

### Momentum Conservation Equation

For momentum conservation, $\mathbf{b} = \mathbf{u}$ and the fluid
velocity is the convective velocity, $\mathbf{v}=\mathbf{u}$. The
generation term is given by
$\rho\mathbf{g}=-\nabla P + \rho \mathbf{G}$. The flux term is given by
$\boldsymbol{\varphi}= -\mu\left(\nabla \mathbf{u} + (\nabla \mathbf{u})^{\mathrm{T}}\right)$.
Thus, we have,

$$\ \displaystyle \frac{\displaystyle \partial^{}(\rho \mathbf{u})}{\displaystyle \partial t ^{}}+\nabla\cdot \left(\rho \mathbf{u}\mathbf{u}\right)-\mu\nabla\cdot \left(\nabla \mathbf{u} + (\nabla \mathbf{u})^{\mathrm{T}}\right)= -\nabla P + \rho \mathbf{G}.$$

For the convective flux term, we can apply the product rule,

$$\ \nabla\cdot \left(\rho\mathbf{u}\mathbf{u}\right)=\nabla\cdot \left(\rho\mathbf{u}\right)\mathbf{u}+\left(\rho\mathbf{u}\right)\cdot\nabla \mathbf{u}.$$

To see why this form of the product rule is true, see Appendix 2. Recall
from the continuity equation,

$$\ 0=\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}}+\nabla\cdot \left(\rho\mathbf{u}\right),$$

$$\ \implies \nabla\cdot \left(\rho\mathbf{u}\right)=-\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}},$$

$$\ \implies \nabla\cdot \left(\rho\mathbf{u}\right)\mathbf{u}=-\mathbf{u}\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}}.$$

Using this, now consider the time derivative and convection terms
together, 

$$\begin{aligned}
\displaystyle \frac{\displaystyle \partial^{}(\rho \mathbf{u})}{\displaystyle \partial t ^{}}+\nabla\cdot \left(\rho\mathbf{u}\mathbf{u}\right)&=\rho\displaystyle \frac{\displaystyle \partial^{}\mathbf{u}}{\displaystyle \partial t ^{}}+\mathbf{u}\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}}-\mathbf{u}\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}}+\rho\mathbf{u}\cdot\nabla\mathbf{u}\\
\ &=\rho\displaystyle \frac{\displaystyle \partial^{}\mathbf{u}}{\displaystyle \partial t ^{}}+\rho\mathbf{u}\cdot\nabla\mathbf{u}\\
\ &=\rho\dfrac{D \mathbf{u}}{Dt}.
\end{aligned}$$ 

The above is valid for compressible and incompressible
flow. Now consider the diffusion term, the divergence of the stress,
which we only developed for incompressible fluids,

$$\ \mu\nabla\cdot \left(\nabla \mathbf{u} + (\nabla \mathbf{u})^{\mathrm{T}}\right)=\mu \nabla^2\mathbf{u}+\mu\nabla\cdot \left(\nabla \mathbf{u}\right)^{\mathrm{T}}.$$

The second term of the above drops out for an incompressible fluid (see
Appendix 2 if interested), leaving only,

$$\ \mathrm{Diffusion}=\mu\nabla^2\mathbf{u}.$$ 

Thus, for an
incompressible fluid, combining all the terms we just developed,

$$\ \underbrace{\rho\displaystyle \frac{\displaystyle \partial^{}\mathbf{u}}{\displaystyle \partial t ^{}}}_{\mathrm{Accumulation}}+\ \ \underbrace{\rho\mathbf{u}\cdot\nabla\mathbf{u}}_{\mathrm{Convection}}\ =\  \underbrace{\mu\nabla^2\mathbf{u}}_{\mathrm{Diffusion}}\ -\ \underbrace{\nabla P+\rho\mathbf{G}}_{\mathrm{Generation}}.$$

Notice that in the convection term, the momentum is convected by its own
movement; this non-linear term makes these equations *extremely* hard to
solve. Therefore, when doing solutions by hand, we often try to find
arguments to get rid of this term such as through scaling analysis.

### Energy Conservation Equation

As a warning, this is a longer derivation with a lot of assumptions and
mathematical trickery to simplify the equation. To start, our conserved
intensive quantity is the total energy of the system, $e$,

$$\ \mathbf{b}=e=u+\frac{1}{2}\mathbf{v}\cdot\mathbf{v}+\mathbf{G}\cdot\mathbf{X},$$

where $u$ is the internal energy, $\mathbf{v}$ is the velocity of medium
(kinetic energy), $\mathbf{G}$ is the gravitational force, and
$\mathbf{X}$ is the position vector of the medium; in total
$\mathbf{G}\cdot\mathbf{X}$ simply represents the gravitational
potential energy. For a constant volume process or an ideal gas,

$$\ du=C_vdT,$$ 

where $C_v$ is the constant volume heat capacity.
Furthermore, for a liquid or solid,
$C_v\approx C_p\approx \mathrm{constant}$. Thus, $u$ may be replaced by
$C_p\,T$ in the following equations for a liquid or solid.

The generation term becomes the heat added to the system and the flow
and shear work done by the system. Mathematically,

$$\ \rho\mathbf{g}=\underbrace{Q}_{\mathrm{Heat}}\ -\ \underbrace{\nabla\cdot \left(\mathbf{v} P\right)}_{\mathrm{Flow\ Work}}\ +\ \underbrace{\Phi+\mu\mathbf{u}\cdot\nabla^2\mathbf{u}}_{\mathrm{Shear\ Work}}.$$

The non-convective flux term comes from Fourier's Law,

$$\ \boldsymbol{\varphi}=\mathbf{q}=-k\nabla T.$$ 

Putting it all together,

$$\ \displaystyle \frac{\displaystyle \partial^{}(\rho e)}{\displaystyle \partial t ^{}} + \nabla\cdot \left(\rho\mathbf{v} e\right)-\nabla\cdot \left(k\nabla T\right)=Q-\nabla\cdot \left(\mathbf{v} P\right)+\Phi+\mu\mathbf{v}\cdot\nabla^2\mathbf{v}.$$

Consider the accumulation and convection terms together,

$$\begin{aligned}
    \displaystyle \frac{\displaystyle \partial^{}(\rho e)}{\displaystyle \partial t ^{}} + \nabla\cdot \left(\rho\mathbf{v} e\right)&=\rho\displaystyle \frac{\displaystyle \partial^{}e}{\displaystyle \partial t ^{}}+e\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}} +e\nabla\cdot (\rho\mathbf{v})+(\rho\mathbf{v})\cdot\nabla e\\
    &=\rho\displaystyle \frac{\displaystyle \partial^{}e}{\displaystyle \partial t ^{}}+\rho\mathbf{v}\cdot\nabla e +e\left(\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}}+\nabla\cdot (\rho\mathbf{v})\right).
\end{aligned}$$ 

The second term is 0 by the continuity equation, thus,

$$\ \displaystyle \frac{\displaystyle \partial^{}(\rho e)}{\displaystyle \partial t ^{}} + \nabla\cdot \left(\rho\mathbf{v} e\right) =\rho\displaystyle \frac{\displaystyle \partial^{}e}{\displaystyle \partial t ^{}}+\rho\mathbf{v}\cdot\nabla e=\rho\dfrac{D e}{Dt}.$$

This is valid for an both compressible and incompressible flows, but the
following developments only consider an incompressible and Newtonian
fluid. First considering the time derivative, 

$$\begin{aligned}
    \rho\displaystyle \frac{\displaystyle \partial^{}e}{\displaystyle \partial t ^{}}&=\rho\left(C_p \displaystyle \frac{\displaystyle \partial^{}T}{\displaystyle \partial t ^{}}+\frac{1}{2}\displaystyle \frac{\displaystyle \partial^{}}{\displaystyle \partial t ^{}}\left(\mathbf{v}\cdot\mathbf{v}\right)+\displaystyle \frac{\displaystyle \partial^{}}{\displaystyle \partial t ^{}}\left(\mathbf{G}\cdot\mathbf{X}\right)\right)\\
    &=\rho C_p \displaystyle \frac{\displaystyle \partial^{}T}{\displaystyle \partial t ^{}} + \rho\mathbf{v}\cdot\displaystyle \frac{\displaystyle \partial^{}\mathbf{v}}{\displaystyle \partial t ^{}}+\rho\mathbf{v}\cdot\mathbf{G},
\end{aligned}$$ 

where the kinetic energy term comes from application of
the product rule and the potential energy term comes from the fact that
$\mathbf{G}$ is a constant vector and the time derivative of the
position vector, $\mathbf{X}$, is the velocity vector, $\mathbf{v}$. Now
for the convection term, 

$$\begin{aligned}
    \rho\mathbf{v}\cdot\nabla e &= \rho C_p \mathbf{v}\cdot\nabla T + \frac{1}{2}\rho\mathbf{v}\cdot\nabla \left(\mathbf{v}\cdot\mathbf{v}\right) + \rho\mathbf{v}\cdot\nabla \left(\mathbf{G}\cdot\mathbf{X}\right)\\
    &=\rho C_p \mathbf{v}\cdot\nabla T + \rho\mathbf{v}\cdot\mathbf{v}\cdot\nabla\mathbf{v}+\rho\mathbf{v}\cdot\mathbf{G}\cdot\nabla\mathbf{X},
\end{aligned}$$ 

where the kinetic energy term comes from the product
rule and and potential energy term arises because $\mathbf{G}$ is a
constant vector. By recombining the accumulation and convection terms,

$$\ \displaystyle \frac{\displaystyle \partial^{}(\rho e)}{\displaystyle \partial t ^{}} + \nabla\cdot \left(\rho\mathbf{v} e\right)=\rho C_p \left(\displaystyle \frac{\displaystyle \partial^{}T}{\displaystyle \partial t ^{}}+\mathbf{v}\cdot\nabla T\right) + \rho\mathbf{v}\cdot\left(\displaystyle \frac{\displaystyle \partial^{}\mathbf{v}}{\displaystyle \partial t ^{}}+\mathbf{v}\cdot\nabla\mathbf{v}\right)+\rho\mathbf{v}\cdot\mathbf{G} +\rho\mathbf{v}\cdot\mathbf{G}\cdot\nabla\mathbf{X}.$$

By the Navier-Stokes equations, we have,

$$\ \rho\mathbf{v}\cdot\left(\displaystyle \frac{\displaystyle \partial^{}\mathbf{v}}{\displaystyle \partial t ^{}}+\mathbf{v}\cdot\nabla\mathbf{v}\right)=\mathbf{v}\cdot\mu\nabla^2\mathbf{v}-\mathbf{v}\cdot\nabla P+\rho\mathbf{v}\cdot\mathbf{G}.$$

This allows us to rewrite the accumulation and convection terms as,

$$\ \displaystyle \frac{\displaystyle \partial^{}(\rho e)}{\displaystyle \partial t ^{}} + \nabla\cdot \left(\rho\mathbf{v} e\right)=\rho C_p\left(\displaystyle \frac{\displaystyle \partial^{}T}{\displaystyle \partial t ^{}}+\mathbf{v}\cdot\nabla T\right)+ \mathbf{v}\cdot\mu\nabla^2\mathbf{v}-\mathbf{v}\cdot\nabla P+2\rho\mathbf{v}\cdot\mathbf{G}+\rho\mathbf{v}\cdot\mathbf{G}\cdot\nabla\mathbf{X}.$$

Plugging this back into the main conservation equation,

$$\ \rho C_p \left(\displaystyle \frac{\displaystyle \partial^{}T}{\displaystyle \partial t ^{}}+ \mathbf{v}\cdot\nabla T\right) + {\mu\mathbf{v}\cdot\nabla^2\mathbf{v}}-{\mathbf{v}\cdot\nabla P}+2\rho\mathbf{v}\cdot\mathbf{G}+\rho\mathbf{v}\cdot\mathbf{G}\cdot\nabla\mathbf{X} -\nabla\cdot \left(k\nabla T\right)$$

$$\ =Q-{\mathbf{v}\cdot\nabla{P}}+\Phi+{\mu\mathbf{v}\cdot\nabla^2\mathbf{v}}.$$
Assuming constant $k$ and cancelling terms,

$$\ \rho C_p\left( \displaystyle \frac{\displaystyle \partial^{}T}{\displaystyle \partial t ^{}}+\mathbf{v}\cdot\nabla T\right) + \left(2\rho\mathbf{v}\cdot\mathbf{G}+\rho\mathbf{v}\cdot\mathbf{G}\cdot\nabla\mathbf{X}\right)= k\nabla^2T +Q+\Phi.$$

Assuming that the fluid does not gain a considerable amount of
gravitational potential energy, we take $\mathbf{G}=\mathbf{0}$, giving,

$$\ \underbrace{\rho C_p\displaystyle \frac{\displaystyle \partial^{}T}{\displaystyle \partial t ^{}}}_{\mathrm{Accumulation}}+\ \ \underbrace{\rho C_p\mathbf{v}\cdot\nabla T}_{\mathrm{Convection}}\ =\  \underbrace{k\nabla^2T}_{\mathrm{Diffusion}}\ \ +\ \underbrace{Q+\Phi}_{\mathrm{Generation}}.$$

This is a practical form of the energy equation for an incompressible
and Newtonian liquid that gains negligible gravitational energy. To
describe heat conduction in a solid, take $\mathbf{v}=\mathbf{0}$ and
$\Phi=0$. To describe an ideal gas that is approximately incompressible,
replace $C_p$ with $C_v$.

## Multicomponent Mass Transfer

### Component Transport Equations

This section is addendum to the mass conservation equation to account
for mass transfer when multiple components are involved. Implicitly, we
only considered one species in a solution or mixture and arrived at the
equation,

$$\ \displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}} + \nabla\cdot \left(\mathbf{v}C\right) = D\nabla^2 C + R,$$

where we assumed a constant diffusion coefficient, $D$. However, this
equation generalizes to each component in the system. Physically,
nothing changes, however, the mathematical notation becomes much more
delicate. Suppose we have an $N$ component system and we consider the
$i$th species. By the general conservation equation for the $i$th
species, we can conclude,

$$\ \displaystyle \frac{\displaystyle \partial^{}C_i}{\displaystyle \partial t ^{}} + \nabla \cdot\left(\mathbf{v}C_i\right) = D_i\nabla^2 C_i + R_i,$$

where $C_i$ is the concentration of the $i$th species, $D_i$ is the
diffusion constant of the $i$th species, and $R_i$ is the generation or
consumption of the $i$th species. If we sum over all of the $N$ species,
we find, $$\ C = \sum_{i=1}^NC_i, \quad R = \sum_{i=1}^N R_i,$$ where
$C$ is the total concentration of the system and $R$ is the total
generation of the system. Thus, the system conservation equation
becomes,

$$\ \displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}} + \sum_{i=1}^N\left(\nabla\cdot \left(\mathbf{v}C_i\right)\right) = \sum_{i=1}^N\left(D_i\nabla^2 C_i\right)+R,$$

or equivalently,

$$\ \displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}} + \nabla\cdot\left(\mathbf{v}\sum_{i=1}^NC_i\right) = \sum_{i=1}^N\left(D_i\nabla^2 C_i\right)+R,$$

thus,

$$\ \displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}} + \nabla\cdot\left(\mathbf{v}C\right) = \sum_{i=1}^N\left(D_i\nabla^2 C_i\right)+R.$$

This is a nice relation, but it doesn't tell us anything new. We'll see
soon that we can simplify the above to get rid of the last summation. A
better form of the above equation comes from considering the mass
(density) of each component instead of the concentration. Using the
relation, $\rho_i=C_i M_i$, where $M_i$ is the molar mass of the $i$th
component, we have,

$$\ \frac{1}{M_i}\displaystyle \frac{\displaystyle \partial^{}\rho_i}{\displaystyle \partial t ^{}} + \frac{1}{M_i}\nabla \cdot\left(\rho_i\mathbf{v}\right) = \frac{D_i}{M_i}\nabla^2 \rho_i + R_i,$$

or equivalently,

$$\ \displaystyle \frac{\displaystyle \partial^{}\rho_i}{\displaystyle \partial t ^{}} + \nabla \cdot \left(\rho_i\mathbf{v}\right) = D_i\nabla^2 \rho_i + M_iR_i.$$

We again sum over all components to find,

$$\ \rho = \sum_{i=1}^N\rho_i, \quad MR = \sum_{i=1}^N M_iR_i,$$ 

where $\rho$ is the total density and $MR$ is the total mass generation of the
system. For most processes, mass is not generated and thus, $MR=0$,
giving,

$$\ \displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}} + \sum_{i=1}^N\left(\nabla\cdot \left(\rho_i\mathbf{v}\right)\right) = \sum_{i=1}^N\left(D_i\nabla^2 \rho_i\right).$$

or,

$$\ \displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}} + \nabla\cdot\left(\mathbf{v}\sum_{i=1}^N\rho_i\right) =  \sum_{i=1}^N\left(D_i\nabla^2 \rho_i\right),$$

which becomes,

$$\ \displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}}+\nabla\cdot \left(\rho\mathbf{v}\right) = \sum_{i=1}^N\left(D_i\nabla^2 \rho_i\right).$$

However, since we are analyzing the total mass of the system, we can
apply the continuity equation to the total system,

$$\ 0=\displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}}+\nabla\cdot \left(\rho\mathbf{v}\right).$$

Thus, we have, 

$$\ \sum_{i=1}^N\left(D_i\nabla^2 \rho_i\right)=0.$$

Physically, this means that the total mass diffusion in a system is 0.
That is, diffusion in one component will be made up for by diffusion in
another component.

### Common Notations for Multicomponent Systems

In addition to the differential equations, there is some important (and
confusing) notation to become accustomed to. Often, we define total the
concentration flux, $\mathbf{N}_i$, as the sum of the non-convective and
convective flux,

$$\ \mathbf{N}_i := \mathbf{J}_i + \mathbf{v}C_i := \mathbf{v}_iC_i,$$

where $\mathbf{J}_i$ comes from Fick's Law of Diffusion,

$$\ \mathbf{J}_i = -D_i\nabla C_i,$$ 

and we have defined the average
component velocity, $\mathbf{v}_i$. We can rearrange the above equations
as, 

$$\ -D_i\nabla C_i = C_i\left(\mathbf{v}_i-\mathbf{v}\right),$$ 

if we consider a stationary fluid with $\mathbf{v}=\mathbf{0}$, then,

$$\ -D_i\nabla C_i = C_i\mathbf{v}_i.$$ 

Physically, this tells us that
diffusion is due to microscopic transport of particles with component
velocity $\mathbf{v}_i$. This tells us that an individual component of
solution can diffuse with a velocity $\mathbf{v}_i$ while the solution
as a whole remains stationary. If we further define the bulk velocity,
$\mathbf{v}$, as,

$$\ \mathbf{v}C := \sum_{i=1}^N \mathbf{v}_iC_i,$$ 

then we can sum over the component flux relations to get,

$$\ \sum_{i=1}^N \mathbf{v}_iC_i = -\sum_{i=1}^N \left(D_i\nabla C_i\right) + \mathbf{v}\sum_{i=1}^N C_i,$$

giving,

$$\ \mathbf{v}C = -\sum_{i=1}^N \left(D_i\nabla C_i\right) + \mathbf{v}C,$$

or, 
$$\ \sum_{i=1}^N \left(D_i\nabla C_i\right)=0.$$ 

If we plug this
back into the summed concentration equation,

$$\ \displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}} + \nabla\cdot\sum_{i=1}^N \left(\mathbf{v}_iC_i\right) = R,$$

or written in terms of the total component fluxes, $\mathbf{N}_i$,

$$\ \displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}} + \nabla\cdot\sum_{i=1}^N \mathbf{N}_i = R,$$

or even more simply,

$$\ \displaystyle \frac{\displaystyle \partial^{}C}{\displaystyle \partial t ^{}} + \nabla\cdot\left(\mathbf{v}C\right) = R.$$

We can also express any concentrations in terms of mole fractions, $x_i$
for a liquid and $y_i$ for a vapor. For example, for a liquid,

$$\ x_i=\frac{C_i}{C},$$ 

with the condition that over all $N$
components, 

$$\ \sum_{i=1}^N x_i = 1.$$ 

To summarize all the notation,

$$\begin{aligned}
\mathbf{N}_i =  \mathbf{v}_iC_i &= \mathbf{J}_i + \mathbf{v}C_i\\ 
&= -D_i\nabla C_i + \mathbf{v}C_i \\
&=-D_i\nabla C_i + x_i\sum_{i=1}^N \mathbf{v}_iC_i\\
&=-D_i\nabla C_i + x_i\sum_{i=1}^N \mathbf{N}_i.
\end{aligned}$$ 

When we consider the density instead of the
concentration, we define the total mass flux, $\mathbf{n}_i$,

$$\ \mathbf{n}_i := \mathbf{j}_i + \rho_i\mathbf{v} := \rho_i\mathbf{v}_i,$$

where, 

$$\ \mathbf{j}_i = -D_i\nabla \rho_i.$$ 

By the same arguments
used in the concentration equation, we find that for the entire system,

$$\ \displaystyle \frac{\displaystyle \partial^{}\rho}{\displaystyle \partial t ^{}} + \nabla\cdot\sum_{i=1}^N\mathbf{n}_i= 0,$$

where the reaction term is 0 since the total mass is conserved. Again,
we can express any of the mass relations this in terms of mass
fractions, $w_i$, $$\ w_i=\frac{\rho_i}{\rho},$$ with the condition that
over all $N$ components, 

$$\ \sum_{i=1}^N w_i = 1.$$ 

To summarize all of the notation, 

$$\begin{aligned}
\mathbf{n}_i =  \rho_i\mathbf{v}_i &= \mathbf{j}_i + \rho_i\mathbf{v}\\ 
&= -D_i\nabla \rho_i + \rho_i\mathbf{v} \\
&=-D_i\nabla \rho_i + w_i\sum_{i=1}^N \rho_i\mathbf{v}_i\\
&=-D_i\nabla \rho_i + w_i\sum_{i=1}^N \mathbf{n}_i.
\end{aligned}$$
