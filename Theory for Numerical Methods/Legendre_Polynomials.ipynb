# Legendre's Differential Equation
Legendre's differential equation is given as follows,

$$\ \left(1-x^2\right)\displaystyle \frac{d^{2}y}{dx^{2}}-2x\displaystyle \frac{d^{}y}{dx^{}}+\lambda y = 0,$$

where $\lambda$ is a constant. We may initially guess that $y$ may be
expanded in terms of a power series,

$$\ y = \sum_{n=0}^{\infty}a_nx^n,$$ 

where the $a_n$ are to be
determined. Plugging this guess into the ODE,

$$\ 0=\displaystyle \frac{d^{2}}{dx^{2}}\sum_{n=0}^{\infty}a_nx^n-x^2\displaystyle \frac{d^{2}}{dx^{2}}\sum_{n=0}^{\infty}a_nx^n-2x\displaystyle \frac{d^{}}{dx^{}}\sum_{n=0}^{\infty}a_nx^n+\lambda\sum_{n=0}^{\infty}a_nx^n$$

$$\ =\sum_{n=2}^{\infty}n(n-1)a_nx^{n-2}-\sum_{n=2}^{\infty}n(n-1)a_nx^n-\sum_{n=1}^{\infty}2na_nx^n+\sum_{n=0}^{\infty}\lambda a_nx^n.$$

For the first sum, reindex the series such that $n-2\to n$, this gives

$$\ 0=\sum_{n=0}^{\infty}(n+1)(n+2)a_{n+2}x^n-\sum_{n=2}^{\infty}n(n-1)a_nx^n-\sum_{n=1}^{\infty}2na_nx^n+\sum_{n=0}^{\infty}\lambda a_nx^n$$

$$\ =2a_0+6a_1x-2a_1+\lambda (a_0+a_1x)+\sum_{n=2}^{\infty}\Bigl[(n+1)(n+2)a_{n+2}-n(n-1)a_n-2na_n+\lambda a_n\Bigr]x^n.$$

For the above series to be identically 0, all of the coefficients of the
series must be 0, thus,

$$\ 0=(n+1)(n+2)a_{n+2}-n(n-1)a_n-2na_n+\lambda a_n,$$

$$\ \implies a_{n+2}=\frac{n(n+1)-\lambda}{(n+1)(n+2)}a_n.$$ 

To find the
coefficients, looking at the even terms, 

$$\begin{aligned}
    &a_0=a_0\\
    &a_2=\frac{(1-\lambda)}{(1)(2)}a_0\\
    &a_4=\frac{(1-\lambda)(6-\lambda)}{(1)(2)(3)(4)}a_0\\
    &a_6=\frac{(1-\lambda)(6-\lambda)(20-\lambda)}{(1)(2)(3)(4)(5)(6)}a_0\\
    &a_n=\frac{(1-\lambda)\cdots(n(n+1)-\lambda)}{n!}a_0
\end{aligned}$$ 

for the odd terms, 

$$\begin{aligned}
    &a_1=a_1\\
    &a_3=\frac{(2-\lambda)}{(2)(3)}a_1\\
    &a_5=\frac{(2-\lambda)(12-\lambda)}{(2)(3)(4)(5)}a_1\\
    &a_7=\frac{(2-\lambda)(12-\lambda)(30-\lambda)}{(2)(3)(4)(5)(6)(7)}a_1\\
    &a_n=\frac{(2-\lambda)\cdots(n(n+1)-\lambda)}{n!}a_1
\end{aligned}$$ 

Looking at the ratio,

$$\frac{a_{n+2}}{a_n}=\frac{n(n+1)-\lambda}{(n+1)(n+2)},$$

as
$n\to\infty$, the ratio approaches 1, thus, solving for the radius of
convergence of the series,

$$\ \lim_{n\to\infty}\left|\frac{a_{n+2}x^{n+2}}{a_nx^n}\right|=\lim_{n\to\infty}x^2.$$

Thus, the series converges for $x\in(-1,1)$, however, we often want
$x\in[-1,1]$, thus, we choose $\lambda$ such that the power series
terminates after a finite amount of terms. Let $\lambda=k(k+1)$, then,


$$\ a_{n+2}=\frac{n(n+1)-k(k+1)}{(n+1)(n+2)}a_n$$

$$\ =\frac{n^2-k^2+n-k}{(n+1)(n+2)}a_n$$

$$\ =-\frac{(n+k+1)(k-n)}{(n+1)(n+2)}a_n,$$ and,

$$\ a_{2n}=(-1)^n\frac{(k+1)(k+3)\cdots(k+2n-1)(k)(k-2)\cdots(k-2n+2)}{(2n)!}a_0,$$

$$\ =\frac{(-1)^n}{(2n)!}\frac{(k+2n-1)!!}{(k-1)!!}\frac{k!!}{(k-2n)!!}a_0$$

$$\ a_{2n+1}=(-1)^n\frac{(k+2)(k+4)\cdots(k+2n)(k-1)(k-3)\cdots(k-2n+1)}{(2n+1)!}a_1$$

$$\ =\frac{(-1)^n}{(2n+1)!}\frac{(k+2n)!!}{k!!}\frac{(k-1)!!}{(k-2n-1)!!}a_1.$$

To get more explicit representations of the coefficients, let $k=2\ell$
for the even series and $k=2\ell+1$ for the odd series. Also consider
the following identities for the double factorial: 

$$\ (2k)!!=2^kk!,$$

$$\ (2k+1)!!=\frac{(2k+1)!}{2^kk!}.$$ Thus,

$$\ a_{2n}=\frac{(-1)^n}{(2n)!}\frac{(2\ell+2n+1)!!}{(2\ell+1)!!}\frac{(2\ell)!!}{(2\ell-2n)!!}a_0$$

$$\ =\frac{(-1)^n}{(2n)!} \frac{(2\ell+2n)!}{2^{\ell+n}(\ell+n)!} \frac{2^\ell\ell!}{(2\ell)!} \frac{2^{\ell+n}(\ell+n)!}{2^{\ell-n}(\ell-n)!}a_0$$

$$\ =(-1)^n\frac{(\ell!)^2(2\ell+2n)!}{(2\ell)!(2n)!(\ell-n)!(\ell+n)!},$$

and,

$$\ a_{2n+1}=\frac{(-1)^n}{(2n+1)!}\frac{(2\ell+2n+1)!!}{(2\ell+1)!!}\frac{(2\ell)!!}{(2\ell-2n)!!}a_1$$

$$\ =\frac{(-1)^n}{(2n+1)!}\frac{(2\ell+2n+1)!}{2^{\ell+n}(\ell+n)!}\frac{2^\ell\ell!}{(2\ell+1)!}\frac{2^\ell\ell!}{2^{\ell-n}
(\ell-n)!}a_1$$

$$\ =(-1)^n\frac{(\ell!)^2(2\ell+2n+1)!}{(2\ell+1)!(2n+1)!(\ell-n)!(\ell+n)!}a_1.$$

With work, it can be shown that both of these coefficients match the
following definition,

$$\ P_\ell(x)=\sum_{n=0}^{\lfloor\ell/2\rfloor}\frac{(-1)^n(2\ell-2n)!}{2^\ell n!(\ell-n)!(\ell-2n)!}x^{\ell-2n}.$$

The following code is a simple implementation of the recursive formula for the coefficients for the Legendre polynomials. It is customary to normalize the polynomials such that $P_n(1)=1$ which is done by dividing the recursive coefficients their sum.

```python
import numpy as np

# Order n Legendre polynomials given input x (vector), n (integer)
def leg_poly(x,n):
    
    max_index = int(np.floor(n/2))
    coeffs = np.ones(max_index+1)
    idx = np.arange(n%2,n,2)
    
    # Use recursive formula for coefficients
    for i in np.arange(0,max_index):
        
        coeffs[i+1] = -coeffs[i]*(n+idx[i]+1)*(n-idx[i])/(idx[i]**2+3*idx[i]+2)
    
    # Normalize coefficients such that P(1) = 1
    coeffs /= np.sum(coeffs)
    
    # Evaluate P(x) at given x points
    poly = np.zeros(x.shape)          # Initialize poly
    idx = np.arange(n%2,n+2,2)        # Polynomial powers
    
    # Sum coeff*powers of x
    for i in np.arange(0,max_index+1):

        poly += coeffs[i]*x**idx[i]
    
    return (poly)
```

```python
import matplotlib.pyplot as plt

# PLot over [-1,1]
x = np.linspace(-1,1,100)
# Highest order to plot
n = 10

# Plot each polynomial
for i in np.arange(0,n):

    P = leg_poly(x,i)
    plt.plot(x,P,label='$P_{' + str(i) + '}(x)$')
    
# Plot settings
plt.legend(bbox_to_anchor=[1,1])
plt.title('First ' + str(n) + ' Legendre Polynomials')
plt.xlabel('$x$')
plt.ylabel('$P_n(x)$')
plt.show()
```

From the explicit relation, it is possible to derive Rodrigues' formula. First
notice,

$$\ \frac{(2\ell-2n)!}{(\ell-2n)!}x^{2\ell-2n}=\displaystyle \frac{d^{\ell}}{dx^{\ell}}x^{2\ell-2n},$$

$$\ \frac{1}{n!(\ell-n)!}=\frac{1}{\ell!}\binom{\ell}{n}.$$ Thus,

$$\ P_\ell=\frac{1}{2^\ell\ell!}\displaystyle \frac{d^{\ell}}{dx^{\ell}}\sum_{n=0}^{\ell}\binom{\ell}{n}(-1)^n(x^2)^{\ell-n},$$

$$\ =\frac{1}{2^\ell\ell!}\displaystyle \frac{d^{\ell}}{dx^{\ell}}\left(x^2-1\right)^\ell.$$

It is possible to prove the orthogonality of the Legendre polynomials
from the above formula, but an easier way is to use the differential
equation,

$$\ \left(1-x^2\right)\displaystyle \frac{d^{2}P_n}{dx^{2}}-2x\displaystyle \frac{d^{}P_n}{dx^{}}+n(n+1) P_n = 0,$$

$$\ \left(1-x^2\right)\displaystyle \frac{d^{2}P_m}{dx^{2}}-2x\displaystyle \frac{d^{}P_m}{dx^{}}+m(m+1) P_m = 0,$$

multiply the first equation by $P_m$ and the second by $P_n$, then
subtract the equations. This gives,

$$\ \left(1-x^2\right)\displaystyle \frac{d^{2}P_n}{dx^{2}}P_m-2x\displaystyle \frac{d^{}P_n}{dx^{}}P_m+2x\displaystyle \frac{d^{}P_m}{dx^{}}P_n-\left(1-x^2\right)\displaystyle \frac{d^{2}P_m}{dx^{2}}P_n+[n(n+1)-m(m+1)]P_nP_m = 0.$$

Notice,

$$\ \displaystyle \frac{d^{}}{dx^{}}\left(\left(1-x^2\right)\left(\displaystyle \frac{d^{}P_n}{dx^{}}P_m-\displaystyle \frac{d^{}P_m}{dx^{}}P_n\right)\right)=\left(1-x^2\right)\left(\displaystyle \frac{d^{2}P_n}{dx^{2}}P_m-\displaystyle \frac{d^{2}P_m}{dx^{2}}P_n\right)-2x\left(\displaystyle \frac{d^{}P_n}{dx^{}}P_m-\displaystyle \frac{d^{}P_m}{dx^{}}P_n\right).$$

Using this to integrate the original equation over $x\in[-1,1]$,

$$\ \left(1-x^2\right)\left(\displaystyle \frac{d^{}P_n}{dx^{}}P_m-\displaystyle \frac{d^{}P_m}
{dx^{}}P_n\right)\Biggl|_{-1}^{1}\ +\ [n(n+1)-m(m+1)]\int_{-1}^{1}P_nP_mdx=0.$$

The boundary term vanishes due to the $1-x^2$ term. This implies,

$$\ [n(n+1)-m(m+1)]\int_{-1}^{1}P_nP_mdx=0,$$

if $m \neq n$, then the
integral must be 0 and the polynomials are orthogonal. If $m=n$, then
the integral is non-zero,

$$\ \int_{-1}^{1}\left(P_n\right)^2dx=\frac{1}{4^n(n!)^2}\int_{-1}^{1}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}dx.$$

Using integration by parts, 

$$\begin{aligned}
\int_{-1}^{1}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}dx&={\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n-1}}{dx^{n-1}}\left(x^2-1\right)^n}\Bigl|_{-1}^{1}\\&-\int_{-1}^{1}{\displaystyle \frac{d^{n-1}}{dx^{n-1}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n+1}}{dx^{n+1}}\left(x^2-1\right)^n}dx. 
\end{aligned}$$ 

The boundary term vanishes using properties of even and
odd functions, giving,

$$\ \int_{-1}^{1}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}dx=-\int_{-1}^{1}{\displaystyle \frac{d^{n-1}}{dx^{n-1}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n+1}}{dx^{n+1}}\left(x^2-1\right)^n}dx.$$

Applying integration by parts $n-1$ more times,

$$\ \int_{-1}^{1}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-
1\right)^n}dx=(-1)^n\int_{-1}^{1}\left(x^2-1\right)^n{\displaystyle \frac{d^{2n}}{dx^{2n}}\left(x^2-1\right)^n}dx.$$

The $2n^{th}$ derivative of $(x^2-1)^n$ is a constant, $(2n)!$, thus,

$$\ \int_{-1}^{1}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}{\displaystyle \frac{d^{n}}{dx^{n}}\left(x^2-1\right)^n}dx=(-1)^n(2n)!\int_{-1}^{1}\left(x^2-1\right)^ndx,$$

giving,

$$\ \int_{-1}^{1}\left(P_n\right)^2dx=\frac{(-1)^n(2n)!}{4^n(n!)^2}\int_{-1}^{1}\left(x^2-1\right)^ndx$$

$$\ =\frac{2(2n)!}{4^n(n!)^2}\int_{0}^{1}\left(1-x^2\right)^ndx.$$ 

For the RHS integral, $I_n$, let $x=\sin{\left(t\right)}$,

$$\ I_n=\int_{0}^{\pi}\left(\cos{\left(t\right)}\right)^{2n+1}dt.$$

Notice,

$$\ \displaystyle \frac{d^{}}{dt^{}}\left(\cos{\left(t\right)}\right)^{2n+1}\sin{\left(t\right)}=-(2n+1)\left(\cos{\left(t\right)}\right)^{2n}\left(\sin{\left(t\right)}\right)^2+\left(\cos{\left(t\right)}\right)^{2n+2}$$

$$\ =(2n+1)\left(\cos{\left(t\right)}\right)^{2n+2}-(2n+1)\left(\cos{\left(t\right)}\right)^{2n}+\left(\cos{\left(t\right)}\right)^{2n+2}$$

$$\ =(2n+2)\left(\cos{\left(t\right)}\right)^{2n+2}-(2n+1)\left(\cos{\left(t\right)}\right)^{2n}.$$

Integrating both sides,

$$\ \left(\cos{\left(t\right)}\right)^{2n+1}\sin{\left(t\right)}\Bigl|_{0}^{\pi}=(2n+2)\int_{0}^{\pi}\left(\cos{\left(t\right)}\right)^{2n+2}dt-(2n+1)\int_{0}^{\pi}\left(\cos{\left(t\right)}\right)^{2n}dt$$

$$\ \implies (2n+2)\int_{0}^{\pi}\left(\cos{\left(t\right)}\right)^{2n+2}dt=(2n+1)\int_{0}^{\pi}\left(\cos{\left(t\right)}\right)^{2n}dt$$

$$\implies (2n+1)I_{n}=2nI_{n-1}$$ 

or,

$$\ I_n=\frac{2n}{2n+1}I_{n-1}.$$

Using $I_0=1$, this relation gives,

$$\ I_n=\frac{(2)(4)\cdots(2n-2)(2n)}{(3)(5)\cdots(2n-1)(2n+1)}$$

$$\ =\frac{(2n)!!}{(2n+1)!!}$$ $$\ =\frac{4^n(n!)^2}{(2n+1)!}.$$

Finally, this gives,

$$\ \int_{-1}^{1}\left(P_n\right)^2dx=\frac{2(2n)!}{4^n(n!)^2}\frac{4^n(n!)^2}{(2n+1)!}$$

$$\ =\frac{2}{2n+1}.$$ Thus, the \"norm\" of the Legendre polynomials is, given by the above. In inner product notation,

$$\ \left\langle P_m,P_n\right\rangle=\frac{2}{2n+1}\delta_{mn}.$$

Moreover, the Legendre polynomials serve as a basis for the interval
$[-1,1]$ (any closed interval can be mapped to this interval).
Symbolically, this is given by, 

$$\ f(x)=\sum_{n=0}^{\infty}a_nP_n(x),$$

where the $a_n$'s are coefficients given by,

$$\ a_n=\left\langle f,P_n\right\rangle=\frac{2n+1}{2}\int_{-1}^{1}f(x)P_n(x)dx.$$

Therefore,

$$\ f(x)=\sum_{n=0}^{\infty}P_n(x)\left(\frac{2n+1}{2}\right)\int_{-1}^{1}f(y)P_n(y)dy,$$

if we formally swap the order of summation and integration,

$$\ f(x)=\int_{-1}^{1}f(y)\Bigl[\sum_{n=0}^{\infty}\left(\frac{2n+1}{2}\right)P_m(y)P_n(x)\Bigr]dy.$$

From this, we can conclude the relation,

$$\ \sum_{n=0}^{\infty}\left(\frac{2n+1}{2}\right)P_n(x)P_n(y)=\delta(x-y),$$

which only holds in the context of the inner product/integral. This is known as the \"completeness relation\" for the Legendre polynomials and is akin to the completeness relation for trigonometric Fourier series.

## Numerical Implementation of the Legendre Polynomials

```python
import seaborn as sns

# Orthogonal interval,x = [-1,1]
x = np.linspace(-1,1,100)

# Grid of x,y values
[X,Y] = np.meshgrid(x,x)

# Delta function approximation using Legendre polynomials
def delta_func(x,y,n_max):
    
    delta = np.zeros(x.shape)
    
    for i in np.arange(0,n_max):
        
        delta += (i + 0.5)*leg_poly(x,i)*leg_poly(y,i)
        
    return delta

# Z(X,Y) values
Z = delta_func(X,Y,25)

# 3D plot of completeness relation
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, linewidth=0, antialiased=False)
plt.xlabel('$x$')
plt.ylabel('$y$')
ax.set_zlim(-15,15)

plt.show()
```

```python
# Heatmap of completeness relation
heatmap = sns.heatmap(Z,vmin=-10,vmax=10)
heatmap.invert_yaxis()
heatmap.set(xticks=[])
heatmap.set(yticks=[])
heatmap.set(xlabel='x',ylabel='y')

plt.show()
```

From these figures, it is clear the sum of polynomials have a spike in the output along the line $y=x$, while there are oscillations about 0 over the rest of the domain. This is a good indication that the completeness relation is being satisfied and the sum is approaching $\delta(x-y)$. For a clearer visual, consider when $y=0$, then we have that the sum should converge to $\delta(x)$.

```python
# Orthogonal interval, x = [-1,1]
x = np.linspace(-1,1,200)

# Order of summation for completeness relation
n_max = 35

# Delta function approximation for given n_max
y = delta_func(x,np.zeros(1),35)

# Plots results
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$\delta(x)$')
plt.title('Completeness Relation of $P_n(x)$ for $n=' + str(n_max) + '$')
plt.show()
```

## Legendre Polynomials as Basis Functions

```python
from scipy.linalg import solve

# Matrix with Legendre Polynomials as columns
def legendre_basis_mtx(x,n):
    
    basis = np.ones([x.size,n])
    
    for i in range(1,n):
        
        basis[:,i] = leg_poly(x,i)
        
    return basis

# Test data for basis function fitting
n = 12
num_data = 20

[x,dx] = np.linspace(-1,1,100,retstep=True)
x_data = np.linspace(-1,1,num_data)

# Normalization constant for normal equations
eps = 1e-16

# Example function to curve fit
data = (np.sin(3*x_data*np.pi)*np.exp(-4*x_data**2) 
        + 0.1*np.random.rand(num_data) - 0.05)

# Basis functions evaluated over [-1,1]
A = legendre_basis_mtx(x_data,n)

# Normal equations for linear least squares
F = A.T@A + eps*np.identity(n)

# Coefficient vector (least squares)
coeffs = solve(F,A.T@data,assume_a='sym')

# Interpolation function
basis_mtx = legendre_basis_mtx(x,n)
interpolant = basis_mtx@coeffs

#--------------------------------------------------------------

# Plot test data
plt.plot(x_data,data,linestyle='none',marker='*',markersize=10)

# Plot interpolant
plt.plot(x,interpolant)

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()

# Print size of coefficients
print('The residual coefficient norm is: %0.2e ' %  
      np.linalg.norm(F@coeffs-A.T@data))

# Print coefficient vector
fig, ax = plt.subplots()
plt.imshow([coeffs,coeffs],cmap='magma')
ax.set_xticks(np.arange(n))
ax.set_yticks([])
plt.colorbar(label='$a_n$',location='bottom')

plt.show()
```

Notice the large oscillations in the interpolation function near $x=-1$ and $x=1$, this is known as '[Runge's phenomenon](https://en.wikipedia.org/wiki/Runge%27s_phenomenon)'.

```python
# Mass matrix between basis functions
idx = np.arange(0,n)
[X,Y] = np.meshgrid(idx,idx)
mass_mtx = basis_mtx.T@basis_mtx*dx

fig, ax = plt.subplots()
plt.pcolormesh(X,Y,mass_mtx,cmap='magma')

ax.set_xticks(idx)
ax.set_yticks(idx)

plt.gca().invert_yaxis()
ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
plt.title('Mass Matrix (Orthogonal Basis)')
plt.colorbar(label='$P_n(x) \cdot P_m(x)$')

plt.show()

print('Mass matrix diagonals: ')
print(np.diagonal(mass_mtx))
```

The mass matrix (Gram matrix) is essentially diagonal, as expected. The diagonals of the matrix should be given by 2/(2n+1). This holds well for the lower order polynomials, but fails for the higher order ones. For example, $P_{11}(x)\cdot P_{11}(x)$ should be about 0.087, but is predicted as 0.116. This deviation is expected as the inner product integral was approximated using only 100 intervals, which is evidently not enough for the higher order polynomials.

```python
# Correlation matrix between coefficients
correlation_mtx = np.outer(coeffs,coeffs)
fig, ax = plt.subplots()
plt.pcolormesh(X,Y,correlation_mtx,cmap='magma')
plt.title('Coefficient Correlation Matrix')
ax.set_xticks(idx[:])
ax.set_yticks(idx[:])
ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
plt.colorbar(label='$a_n \cdot a_m$')
ax.invert_yaxis()

plt.show()
```
